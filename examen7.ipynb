```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cuentas_credicel.csv", encoding="latin")
df.head(5)
```

```python
#Todas las columnas excepto
cols_drop = ["folio", "folio_solicitud","Unnamed: 33", "Unnamed: 34", "Unnamed: 35", "Unnamed: 36"]
df = df.drop(columns=[c for c in cols_drop if c in df.columns])
df.head(5)
```

```python
#todas las filas multiplos de 4
df = df.iloc[::4, :]
df.head(5)
```

```python
#Todos los registros cuya marca sea igual a: "ALCATEL", "MOTOROLA", "NOKIA", "ZTE" y "SAMSUNG"
df["marca"].unique()
filtroc = ["ALCATEL", "MOTOROLA", "NOKIA", "ZTE", "SAMSUNG"]
df = df[df["marca"].isin(filtroc)]
df.head(5)
```

```python
#Todos los registros cuyo enganche se encuentre entre: "$500 y $2000"
df["enganche"] = (df["enganche"].astype(str).str.replace(r"[^\d\.]", "", regex=True).astype(float))
df = df[(df["enganche"] >= 500) & (df["enganche"] <= 2000)]
df.head(5)
```

```python
#Todos los registros cuya categoría en riesgo: NO sea "Cancelado" ni "Fraude"
df = df[~df["riesgo"].isin(["Cancelado", "Fraude"])]
df.head()
```

```python
#Todos los registros cuya fecha : Sea posterior a "15/11/2021"
df["fecha"] = pd.to_datetime(df["fecha"],errors="coerce")
df = df[df["fecha"] > pd.to_datetime("15/11/2021")]
df.head()
```

```python
df.to_csv("evaluacion3.csv")
```
